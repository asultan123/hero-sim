```python
def interpolate_layer_metric(
    layer_properties: pd.DataFrame, metric_type: str, longest_network_length: int
):
    layer_prop_dict = layer_properties.to_dict(orient="index")
    mem_sizes = {}
    for (model_name, _), properties in layer_prop_dict.items():
        size = properties[metric_type]
        try:
            mem_sizes[model_name].append(size)
        except KeyError as e:
            mem_sizes[model_name] = [size]
    interp_sizes = {}
    for model_name, values in mem_sizes.items():
        model_length = len(values)
        if model_length == 1:
            continue
        idx_list = np.linspace(0, longest_network_length, model_length)
        interp = np.interp(
            x = np.arange(longest_network_length), xp = idx_list, fp = values
        )
        interp_sizes[model_name] = interp
    return interp_sizes
```

```python
def get_model_layer_types(model):
    return [
        f'{type(layer)}'.split('.')[-1][:-2]
        for name, layer in model.named_modules()
        if len(list(layer.children())) == 0
    ]


layer_properties = pd.read_csv("../data/layer_metrics.csv", index_col=[0, 1])
model_list = layer_properties.reset_index()['level_0'].drop_duplicates().to_list()

types_counter = Counter()
for model_name in tqdm(model_list):
    model = timm.create_model(model_name, pretrained=False).to('cpu')
    types = get_model_layer_types(model)
    types_counter.update(Counter(types))
    
records = []
for type, freq in dict(types_counter).items():
    records.append({
        'type': type,
        'occurences': freq
    })
layer_type_occurences_df = pd.DataFrame.from_records(records).sort_values('occurences', ascending=False)
```

```python
# psum storage requirement (retiring filters as fast as possible) alap -> keeping entire ofmap on chip asap -> keeping at most f on chip
def compute_psum_trends():
    def load_stats_dict():
        with open('../data/stats_dict.backup', 'rb') as backup:
            stats_dict = pickle.load(backup)
        return stats_dict
    stats_dict = load_stats_dict()

    def compute_psum_storage_wc(filter_parallelism = 64): 
        longest_network_length = max([len(stats['raw_stats'].keys()) for model, stats in stats_dict.items()])
        normalized_psum_wc_per_layer = {}
        for model, stats in stats_dict.items():
            wc_psum_storage_per_layer_asap = []
            wc_psum_storage_per_layer_alap = []
            for layer in stats['raw_stats'].values():
                if layer.kernel_size == (1,1) or layer.kernel_size == (3,3) or layer.kernel_size==(5,5) or layer.kernel_size==(7,7):
                    psum_size = layer.output_size[-1]*layer.output_size[-2]
                    filter_count = layer.output_size[0]
                    max_filters_in_flight_asap = filter_parallelism if filter_count > filter_parallelism else filter_count
                    max_filters_in_flight_alap = filter_count
                    worst_case_storage_for_filters_in_flight_with_asap_tile_sched = psum_size*max_filters_in_flight_asap
                    worst_case_storage_for_filters_in_flight_with_alap_tile_sched = psum_size*max_filters_in_flight_alap

                wc_psum_storage_per_layer_asap.append(worst_case_storage_for_filters_in_flight_with_asap_tile_sched)
                wc_psum_storage_per_layer_alap.append(worst_case_storage_for_filters_in_flight_with_alap_tile_sched)
            model_length = len(stats['raw_stats'])
            idx_list = linspace(0, longest_network_length, model_length)
            interp_wc_psum_asap = np.interp(np.arange(longest_network_length), idx_list, wc_psum_storage_per_layer_asap)
            interp_wc_psum_alap = np.interp(np.arange(longest_network_length), idx_list, wc_psum_storage_per_layer_alap)
            normalized_psum_wc_per_layer[model] = (interp_wc_psum_asap, interp_wc_psum_alap)

        wc_asap = []
        wc_alap = []
        for asap, alap in normalized_psum_wc_per_layer.values():
            wc_asap.extend(asap)
            wc_alap.extend(alap)
        wc_asap = np.array(wc_asap)
        wc_alap = np.array(wc_alap)
        idxs = np.tile(np.arange(longest_network_length), len(normalized_psum_wc_per_layer.keys()))
        idxs = np.array([i/longest_network_length*100 for i in idxs])

        return pd.DataFrame({'idxs':idxs, 'asap':wc_asap, 'alap': wc_alap})

    df = compute_psum_storage_wc(filter_parallelism = 1)
    sns_plt = sns.lineplot(
        data=df, x="idxs", y="alap"
    )
    legend = ['ALAP tile sched']

    for f in [2**i for i in range(0, 10, 2)]:
        print(f'processing {f}')
        df = compute_psum_storage_wc(filter_parallelism = f)
        sns_plt = sns.lineplot(
            data=df, x="idxs", y="asap"
        )
        legend.append(f'ASAP tile sched with f = {f}')
        
    sns_plt.set_yscale('log', base = 2)
    sns_plt.xaxis.set_major_formatter(PercentFormatter())

    sns_plt.set_xlabel('Relative position from start in network')
    sns_plt.set_ylabel('Size in bytes')
    sns_plt.set_title('OFmap storage at different tile schedules')
    sns_plt.legend(legend)
compute_psum_trends()
```

```python
# ifmap storage requirement (retiring filters as fast as possible) alap -> keeping entire ofmap on chip asap -> keeping at most f on chip
def compute_ifmap_trends():
    def load_stats_dict():
        with open('../data/stats_dict.backup', 'rb') as backup:
            stats_dict = pickle.load(backup)
        return stats_dict
    stats_dict = load_stats_dict()

    def compute_ifmap_storage_wc(channel_parallelism = 64, expected_kernel_sizes = [1, 3]): 
        longest_network_length = max([len(stats['raw_stats'].keys()) for model, stats in stats_dict.items()])
        normalized_ifmap_wc_per_layer = {}
        for model, stats in stats_dict.items():
            wc_ifmap_storage_per_layer_asap = []
            wc_ifmap_storage_per_layer_alap = []
            for layer in stats['raw_stats'].values():
                if layer.kernel_size[0] in expected_kernel_sizes and layer.stride[0] == 1:
                    ifmap_size = layer.input_size[-1]*layer.input_size[-2]
                    channel_count = layer.input_size[1]
                else:
                    ifmap_size = layer.input_size[-1]*layer.output_size[1]
                    channel_count = layer.kernel_size[0]*layer.input_size[1]
                max_channels_in_flight_asap = channel_count
                max_channels_in_flight_alap = channel_parallelism if channel_count > channel_parallelism else channel_count
                worst_case_storage_for_channels_in_flight_with_asap_tile_sched = ifmap_size*max_channels_in_flight_asap
                worst_case_storage_for_channels_in_flight_with_alap_tile_sched = ifmap_size*max_channels_in_flight_alap


                wc_ifmap_storage_per_layer_asap.append(worst_case_storage_for_channels_in_flight_with_asap_tile_sched)
                wc_ifmap_storage_per_layer_alap.append(worst_case_storage_for_channels_in_flight_with_alap_tile_sched)
            model_length = len(stats['raw_stats'])
            idx_list = linspace(0, longest_network_length, model_length)
            interp_wc_ifmap_asap = np.interp(np.arange(longest_network_length), idx_list, wc_ifmap_storage_per_layer_asap)
            interp_wc_ifmap_alap = np.interp(np.arange(longest_network_length), idx_list, wc_ifmap_storage_per_layer_alap)
            normalized_ifmap_wc_per_layer[model] = (interp_wc_ifmap_asap, interp_wc_ifmap_alap)

        wc_asap = []
        wc_alap = []
        for asap, alap in normalized_ifmap_wc_per_layer.values():
            wc_asap.extend(asap)
            wc_alap.extend(alap)
        wc_asap = np.array(wc_asap)
        wc_alap = np.array(wc_alap)
        idxs = np.tile(np.arange(longest_network_length), len(normalized_ifmap_wc_per_layer.keys()))
        idxs = np.array([i/longest_network_length*100 for i in idxs])

        return pd.DataFrame({'idxs':idxs, 'asap':wc_asap, 'alap': wc_alap})

    from math import inf
    df = compute_ifmap_storage_wc(channel_parallelism = inf)
    sns_plt = sns.lineplot(
        data=df, x="idxs", y="asap"
    )
    legend = ['ASAP tile sched']

    for c in [2**i for i in range(0, 10, 2)]:
        print(f'processing {c}')
        df = compute_ifmap_storage_wc(channel_parallelism = c)
        sns_plt = sns.lineplot(
            data=df, x="idxs", y="alap"
        )
        legend.append(f'ALAP tile sched with c = {c}')
        
    sns_plt.set_yscale('log', base = 2)
    sns_plt.xaxis.set_major_formatter(PercentFormatter())

    sns_plt.set_xlabel('Relative position from start in network')
    sns_plt.set_ylabel('Size in bytes')
    sns_plt.set_title('IFmap storage at different tile schedules')
    sns_plt.legend(legend)
compute_ifmap_trends()
```
